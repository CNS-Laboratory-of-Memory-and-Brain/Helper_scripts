```python
import pandas as pd
import numpy as np
import json
import os
import shutil
```

```python
count_dict = {}
for subnum in range(10, 55):
    sub = f"sub-{subnum:03d}" 
    path = f"//pentagon.ucsd.edu/squirelab/local/mridata/Smith_Merit/Processed_Data/derivatives/FMRI_Sentences/{sub}/ses-sen/func/"
    if os.path.exists(path):
        f = open(path + f"{sub}_censor.txt", "w")
        to_write = ""
        for file in os.listdir(path):
            if file[-3:] == "tsv" and file[:1] != ".":
                print(f"Working on {file}")
                tsv_path = path + file
                df = pd.read_csv(tsv_path, delimiter='\t')
                col_lst = []
                for col in df.columns:
                    if col[:6] == "motion":
                        col_lst.append(col)
                df['motion_outlier'] = df[col_lst].sum(axis=1)
                df['motion_outlier'] = df['motion_outlier'].apply(lambda count: int(count > 0))
                count_dict[file] = sum(df['motion_outlier'])
                to_write_lst = df['motion_outlier'].apply(lambda count: int(count <= 0)).values
                for value in to_write_lst:
                    to_write += str(value) + "\n"
                    
        f.write(to_write)
        f.close()
```

```python
for subnum in range(10, 55):
    sub = f"sub-{subnum:03d}" 
    path = f"//pentagon.ucsd.edu/squirelab/local/mridata/Smith_Merit/Processed_Data/derivatives/FMRI_Sentences/{sub}/ses-sen/func/"
    if os.path.exists(path):
        if not os.path.exists(path + "confounds/"):
            os.makedirs(path + "confounds/")
        
        
        for col_name in cols:
            file_path = path + "confounds/" + sub + f"_{col_name}.txt"
            print(f"working on {file_path}")
            f = open(file_path, "w")
            to_write = ""
        
            for file in os.listdir(path):
                if file[-3:] == "tsv" and file[:1] != ".":
                    tsv_path = path + file
                    df = pd.read_csv(tsv_path, delimiter='\t')
                    val_array = df[col_name].fillna(0).values
                    for val in val_array:
                        to_write += f"{val}\n"
                            
            f.write(to_write)
            f.close()
```
